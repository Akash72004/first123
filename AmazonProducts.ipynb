# New Section

**Top** **ManuFacturers**

```python
top_ten_manufacturers=df.groupby("manufacturer").count().sort_values(by="uniq_id",ascending=False)["uniq_id"][:10]
top_ten_manufacturers
```

```python
top_ten_manufacturers.plot.barh()
plt.show()
```

**Top 10 Expensive Products**

```python
top_ten_expensive=df[["price","product_name"]].sort_values(by="price", ascending=False)[:10]
top_ten_expensive
```

**Top 10 Cheapest Products**

```python
top_ten_cheap=df[["price","product_name"]].sort_values(by="price", ascending=True)[:10]
top_ten_cheap
```

**Products with Best Reviews**

```python
top_selling_products= df[["number_of_reviews", "product_name" ,"average_review_rating"]].sort_values(by="average_review_rating",ascending=False)[:1000]
top_selling_products
```
